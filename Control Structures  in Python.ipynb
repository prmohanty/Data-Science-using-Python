### Conditional Branching in Python using 'IF - ELSE'

#### IF 

if statement is the most simple decision making statement. It is used to decide whether a certain statement or block of statements will be executed or not i.e if a certain condition is true then a block of statement is executed otherwise not.



```python
# Sample Code using 'IF'

Student = 'Python'

Class = 'Python'

if Student == Class :
    print(" You are in IF Block")
    print( " Welcome Students to the class of Python")

print('\n Out of IF Condition')
```

```python
# Sample Code using 'IF'  ==> Note : The Condition expression can be enclosed in paranthesis

Student = 'Python'

Class = 'Python'

if (Student == Class) :
    print( " Welcome Students to the class of Python")
    
print('\n Out of IF Condition')
```

```python
# Sample Code using 'IF'  ==> When the if Condition is False 

Student = 'Python'

Class = 'R'

if (Student == Class) :
    print( " Welcome Students to the class of Python")   


print('\n Out of IF Condition')
```

```python
# Sample Code using 'IF'  ==> When the if Condition block is more than 1 Python statement

Student = 'Python'

Class = 'Python'

if (Student == Class) :
    print( " The If Condition is True ")
    print( " Welcome Students to the class of Python")
print('\n Out of IF Condition')
```

## IF - ELSE 



```python
# Sample Code using 'IF - ELSE'  : When IF Condition is True

Student = 'Python'

Class = 'Python'

if Student == Class :
    print(" You are in IF Block - ")
    print( " Welcome Students to the class of Python .")
else :
    print(" You are in ELSE Block -")
    print(" Sorry , you are in wrong Class .")
print('\n Out of IF-ELSE Condition')
```

```python
# Sample Code using 'IF - ELSE'  : When ELSE Condition is True

Student = 'Python'

Class = 'R'

if Student == Class :
    print(" You are in IF Block - ")
    print( " Welcome Students to the class of Python .")
else :
    print(" You are in ELSE Block -")
    print(" Sorry , you are in wrong Class .")
print('\n Out of IF-ELSE Condition')
```

### IF - ELIF - ELSE 



```python
# Sample Code using 'IF - ELIF - ELSE'  : When IF Condition is TRUE

Student = 'Python'

Class = 'Python'

if Student == Class :
    print(" You are in IF Block - ")
    print( " Welcome Students to the class of Python .")
elif Student in ('Python' , 'R') :
    print( " You are in ELIF block -")
    print( " You are looking for Data Science topic other than Python ")
else :
    print(" You are in ELSE Block -")
    print(" Sorry , you are looking for non data science topic .")
print('\n Out of IF-ELIF-ELSE Condition')
```

```python
# Sample Code using 'IF - ELIF - ELSE'  : When ELIF condition is True

Student = 'R'

Class = 'Python'

if Student == Class :
    print(" You are in IF Block - ")
    print( " Welcome Students to the class of Python .")
elif Student in ('Python' , 'R') :
    print( " You are in ELIF block -")
    print( " You are looking for Data Science topic other than Python ")
else :
    print(" You are in ELSE Block -")
    print(" Sorry , you are looking for non data science topic .")
print('\n Out of IF-ELIF-ELSE Condition')
```

```python
# Sample Code using 'IF - ELIF - ELSE'  : When ELIF condition is True

Student = 'SAP'

Class = 'Python'

if Student == Class :
    print(" You are in IF Block - ")
    print( " Welcome Students to the class of Python .")
elif Student in ('Python' , 'R') :
    print( " You are in ELIF block -")
    print( " You are looking for Data Science topic other than Python ")
else :
    print(" You are in ELSE Block -")
    print(" Sorry , you are looking for non data science topic .")
print('\n Out of IF-ELIF-ELSE Condition')
```

### WHILE Loop 

In Python, all the statements indented by the same number of character spaces after a programming construct are considered to be part of a single block of code. Python uses indentation as its method of grouping statements.


Note : WHILE Loop is used when you donot know upfront how many iterations to be executed

```python
# Very basic example of WHILE loop 

i = 0 

while i < 5:
    print('Inside the While Loop , Value of i is ',i)
    i += 1   # i = i + 1 

print('\n Outside of WHILE Loop')
```

```python
# Define the Data Structures to be used for WHILE loop 

Training_Dict = {'Matlab':['Manish'],'Python':['Satish','Alex','Eli'],'R':['Janet','Soumya']}

Training_Dict

```

## Loop Control Statements

Loop control statements change execution from its normal sequence.

#### CONTINUE 

It returns the control to the beginning of the loop.

```python
for each elements in a collection :
    Do this 
    Do this 
    Do this
    if evaluate a condition :
        continue 
    else :
        Do this
        Do this
        
```

```python
Training_Dict['Python']
```

```python
# Print all Student names who had opted for Python training except name of Alex

for students in Training_Dict['Python'] :
    if students == 'Alex' :
        continue
    else :
        print('Student Name :',students)
```

### Break Statement

It takes control out of the loop

```python
Training_Dict['Python']
```

```python
# Print all Student names who had opted for Python training before Alex
# Assuming that the enrollment order is maintained in the list

for students in Training_Dict['Python'] :
    if students == 'Alex' :
        break
    else :
        print('Student Name :',students)
```
